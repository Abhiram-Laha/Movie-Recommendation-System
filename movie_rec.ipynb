```python
import pandas as pd
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies =movies.merge(credits,on='title')
```

```python
movies.head(5)
```

```python
# genres
# id
# keywords
# title
# overview
# cast
# crew

movies=movies[['id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head(1)
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
# '[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'

# ['Action','Adventure','Fantasy','Science Fiction']
import ast
```

```python
def convert(obj):
    list=[]
    
    for i in ast.literal_eval(obj):
        list.append(i['name'])
        
    return list

# convert('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
def convert3(obj):
    list=[]
    counter=0
    for i in ast.literal_eval(obj):
        if(counter!=5):
            list.append(i['name'])
            counter+=1
        else:
            break
        
    return list
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies.head(1)
```

```python
def fetch_dict(obj):
    list=[]
    
    for i in ast.literal_eval(obj):
        if(i['job']=='Director'):
            list.append(i['name'])
            break;
        
    return list
```

```python
movies['crew']=movies['crew'].apply(fetch_dict)
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head(1)
```

```python
movies['genres']=movies['genres'].apply(lambda x: [i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x: [i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x: [i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x: [i.replace(" ","") for i in x])
```

```python
movies['tags'] = movies['overview']  + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new_movies = movies[['id','title','tags']]        # new data set with proper data
```

```python
new_movies.head()   
```

```python
new_movies['tags']=new_movies['tags'].apply(lambda x:" ".join(x))
```

```python
new_movies['tags']=new_movies['tags'].apply(lambda x:x.lower())
```

```python
new_movies.head(5)
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
new_movies['tags']=new_movies['tags'].apply(stem)
```

```python
import sklearn.feature_extraction.text as cva

cv = cva.CountVectorizer(max_features=6000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_movies['tags']).toarray()
```

```python
cv.get_feature_names()
```

```python
vectors
```

```python
import sklearn.metrics.pairwise as sm
```

```python
simliarity = sm.cosine_similarity(vectors)
```

```python
def recommend(movie):
    movie_index = new_movies[new_movies['title'] == movie].index[0]
    distance = simliarity[movie_index]
    
    movie_list = sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[1:11]
    a=1
    for i in movie_list:
        print(a," . ",new_movies.iloc[i[0]].title)
        a=a+1
```

```python
# movie recommender

def suggest_movie():
    name=input("Enter Movie Name : ")
    
    print("\nSuggested Movies : ")
    recommend(name)
    
```

```python
recommend('Iron Man')
```
